```python
import pandas as pd
```

```python
df = pd.read_csv('shootings1982_2023.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Renaming columns
df.rename(columns={
    'location.1': 'location_type',
    'gender': 'sex',
    'prior_signs_mental_health_issues': 'mental_health_issues'}, 
    inplace=True)

df.head()
```

```python
# dropping unnecessary columns
# total_victims column had an incorrect amount for the first row, will fix below
df.drop('total_victims', axis=1, inplace=True)
df.drop('latitude', axis=1, inplace=True)
df.drop('longitude', axis=1, inplace=True)
df.drop('where_obtained', axis=1, inplace=True)
df.drop('weapon_details', axis=1, inplace=True)
df.drop('mental_health_details', axis=1, inplace=True)
df.drop('mental_health_issues', axis=1, inplace=True)

df.head(2)
```

```python
# fixing the total column to ensure all totals are correct
df['injured'] = df['injured'].replace(['(TK - "fewer than 10"'], '0')

df['fatalities'] = pd.to_numeric(df['fatalities'], errors='coerce')
df['injured'] = pd.to_numeric(df['injured'], errors='coerce')

df['total_victims'] = df['fatalities'] + df['injured']

total_victims = df.pop('total_victims')
df.insert(6, 'total_victims', total_victims)

df.head()
```

```python
df['sex'] = df['sex'].replace(['F (identified as transgender, per police)'], 'F')
df.loc[df['sex'] == 'Female', 'sex'] = 'F'
df.loc[df['sex'] == 'Male', 'sex'] = 'M'
df.loc[df['sex'] == 'Male & Female', 'sex'] = 'M&F'

df.head(3)
```

```python
df.to_excel('gun_data.xlsx', sheet_name='Data')
```
